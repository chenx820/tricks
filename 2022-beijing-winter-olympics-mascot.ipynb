## 冰墩墩

```python
# Figure
fig = plt.figure()
ax = fig.add_subplot()

color_list=[(98/255,205/255,247/255),(251/255,184/255,77/255),'black',(132/255,199/255,114/255), (158/255,48/255,87/255),] # 奥运五环的颜色

# 画图形；用椭圆堆砌
# 身体
ax.add_patch(Ellipse((0, -0.3), 6+0.3, 6.6+0.3, ec='black', fc='white'))  # Body
ax.add_patch(Ellipse((1.7, 2.4), 1.2+0.3, 1.25+0.3, ec='black', fc='white'))  # Ear
ax.add_patch(Ellipse((-1.7, 2.4), 1.2+0.3, 1.25+0.3, ec='black', fc='white'))

ax.add_patch(Ellipse((-3, -0.9), 2.8+0.3, 1.2+0.3, angle=30, ec='black', fc='white'))  # Left arm
ax.add_patch(Ellipse((-3.7, -1.3), 1.16+0.3, 1.16+0.3, ec='black', fc='white'))
ax.add_patch(Ellipse((-3, -0.9), 2.8+0.2, 1.2+0.2, angle=30, ec='white', fc='white'))
ax.add_patch(Ellipse((3, 0.7), 2.8+0.3, 1.2+0.3, angle=45, ec='black', fc='white'))  # Right arm
ax.add_patch(Ellipse((3.6, 1.2), 1.16+0.3, 1.16+0.3, ec='black', fc='white'))
ax.add_patch(Ellipse((3, 0.7), 2.8+0.2, 1.2+0.2, angle=45, ec='white', fc='white'))

ax.add_patch(Ellipse((1.2, -3.2), 1.6+0.3, 2.4+0.3, ec='black', fc='white'))  # Right leg
ax.add_patch(Ellipse((1.15, -4), 1.4+0.3, 1+0.3, ec='black', fc='white'))
ax.add_patch(Ellipse((1.2, -3.2), 1.6+0.2, 2.4+0.2, ec='white', fc='white'))
ax.add_patch(Ellipse((-1.2, -3.2), 1.6+0.3, 2.4+0.3, ec='black', fc='white')) #  Left leg
ax.add_patch(Ellipse((-1.15, -4), 1.4+0.3, 1+0.3, ec='black', fc='white'))
ax.add_patch(Ellipse((-1.2, -3.2), 1.6+0.2, 2.4+0.2, ec='white', fc='white'))

ax.add_patch(Ellipse((0, -0.3), 6+0.2, 6.6+0.2, ec='white', fc='white'))  # Body

# Ears
ax.add_patch(Ellipse((1.7, 2.4), 1.2, 1.25, ec='black', fc='black'))
ax.add_patch(Ellipse((-1.7, 2.4), 1.2, 1.25, ec='black', fc='black'))

# Arms
ax.add_patch(Ellipse((-3, -0.9), 2.8, 1.2, angle=30, ec='black', fc='black'))
ax.add_patch(Ellipse((-3.7, -1.3), 1.16, 1.16, ec='black', fc='black'))
ax.add_patch(Ellipse((3, 0.7), 2.8, 1.2, angle=45, ec='black', fc='black'))
ax.add_patch(Ellipse((3.6, 1.2), 1.16, 1.16, ec='black', fc='black'))

# Legs
ax.add_patch(Ellipse((1.2, -3.2), 1.6, 2.4, ec='black', fc='black'))
ax.add_patch(Ellipse((1.15, -4), 1.4, 1, ec='black', fc='black'))
ax.add_patch(Ellipse((-1.2, -3.2), 1.6, 2.4, ec='black', fc='black'))
ax.add_patch(Ellipse((-1.15, -4), 1.4, 1, ec='black', fc='black'))

# Body
ax.add_patch(Ellipse((0, -0.3), 6, 6.6, ec='black', fc='white'))

# 脸周围的五环
for i in range(5):
    width = 0.06
    tt = np.linspace(0, np.pi, 100)
    x1 = (2.45-width*i) * np.cos(tt)
    y1 = (2.4-width*i) * np.sin(tt)
    x2 = (2.45-width*i) * np.cos(tt)
    y2 = -(1.7-width*i) * np.sin(tt)
    ax.plot(x1,y1,color=color_list[i],linewidth=1)
    ax.plot(x2,y2,color=color_list[i],linewidth=1)

# Eyes
ax.add_patch(Ellipse((1.1, 1), 1, 1.4, angle=55, ec='black', fc='black'))
ax.add_patch(Ellipse((0.9, 1), 0.5, 0.5, ec='white', fc='black'))
ax.add_patch(Ellipse((1, 1), 0.05, 0.05, ec='white', fc='white'))
ax.add_patch(Ellipse((-1.1, 1), 1, 1.4, angle=-55, ec='black', fc='black'))
ax.add_patch(Ellipse((-0.9, 1), 0.5, 0.5, ec='white', fc='black'))
ax.add_patch(Ellipse((-0.8, 1), 0.05, 0.05, ec='white', fc='white'))

# Mouth
ax.add_patch(Ellipse((0, -0.1), 1.2, 1, angle=20, ec='black', fc='black'))
ax.add_patch(Ellipse((0, 0.4), 0.8, 0.5, angle=10, ec='white', fc='white'))
ax.add_patch(Ellipse((0, -0.28), 0.85, 0.6, angle=10, ec='black', fc=(180/255,39/255,45/255)))

# Nose
ax.add_patch(Ellipse((0, 0.5), 0.4, 0.2, ec='black', fc='black'))

# Rings
tt = np.linspace(0, 2*np.pi, 100)
x = 0.2*np.sin(tt)
y = 0.2*np.cos(tt) - 3.2
ax.plot(x-0.5,y+0.15,color=color_list[0],linewidth=0.65)
ax.plot(x-0.25,y,color=color_list[1],linewidth=0.65)
ax.plot(x,y+0.15,color=color_list[2],linewidth=0.65)
ax.plot(x+0.25,y,color=color_list[3],linewidth=0.65)
ax.plot(x+0.5,y+0.15,color=color_list[4],linewidth=0.65)
    
    # Heart
tt=np.linspace(-2.9,2.9,1000)
x=0.3*(np.sin(tt))**3
y=0.2*(1.3*np.cos(tt)-.5*np.cos(2*tt)-.2*np.cos(3*tt)-.1*np.cos(4*tt))
ax.fill(x+3.6, y+1.2, color=(180/255,39/255,45/255))

# LOGO
ax.text(0, -2.75, 'BEIJING 2022', ha='center', fontsize=5) 

# 设置图形的宽高比为1:1
ax.set_aspect('equal')

# 调整坐标轴范围以包含整个图形
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 4.5)

# 移除坐标轴
plt.axis('off')

# 显示图形
plt.show()
```

## 雪容融

```python
# Figure
fig = plt.figure()
ax = fig.add_subplot()

# Define color
gold = (255/255, 215/255, 0) 
cnred = (230/255, 0, 0)
    
# Hand
ax.add_patch(Ellipse((-1.4, -2.5), 1.65, 0.65, angle=-30, color=cnred))  # Left
ax.add_patch(Ellipse((1.45, -2.9), 1.6, 0.6, angle=-40, color=cnred))  # Right

# Head
ax.add_patch(Ellipse((0, 2), 4.3, 1.7, color=cnred))
ax.add_patch(Ellipse((0, 0), 6.2, 4.8, color=cnred))
ax.add_patch(Ellipse((0, 0), 4.6, 5, ec=gold, fc=cnred))
ax.add_patch(Ellipse((0, 0), 2.6, 5, ec=gold, fc=cnred))
ax.add_patch(Ellipse((0, 0), 0.01, 5, ec=gold, fc=cnred))
ax.add_patch(Ellipse((0, 0), 3.5, 1.8, color='white'))
ax.add_patch(Ellipse((-1.2, -0.5), 1.4, 1, color='white'))
ax.add_patch(Ellipse((-0.6, 0.6), 1.5, 1, color='white'))
ax.add_patch(Ellipse((0, 0.8), 1.5, 1, color='white'))
ax.add_patch(Ellipse((-1.35, -0.68), 0.65, 0.6, color='white'))
ax.add_patch(Ellipse((1.4, -0.5), 1, 0.8, angle=160, color='white'))


# 轮廓
tt = np.linspace(np.pi/2, np.pi/2*3, 100)
x = 3.25 * np.sin(tt)
y = 2.5 * np.cos(tt) + 0.1
ax.plot(x, y, color='w', linewidth=1)
    
tt = np.linspace(-np.pi/3, np.pi/3, 100) # 头上方
x = 3.3 * np.sin(tt)
y = 2.5 * np.cos(tt) + 0.21
ax.plot(x, y, color='w', linewidth=5)
    
# 眼睛 & 腮红
ax.add_patch(Ellipse((1, 0), 0.2, 0.3, ec='black', fc='black'))
ax.add_patch(Ellipse((-1, 0), 0.2, 0.3, ec='black', fc='black'))
ax.add_patch(Ellipse((1, 0.1), 0.02, 0.02, ec='white', fc='white'))
ax.add_patch(Ellipse((-1, 0.1), 0.02, 0.02, ec='white', fc='white'))
ax.add_patch(Ellipse((1.4, -0.45), 0.5, 0.4, alpha=0.4, ec='white', fc=cnred))
ax.add_patch(Ellipse((-1.4, -0.45), 0.5, 0.4, alpha=0.4, ec='white', fc=cnred))

# Legs
ax.add_patch(Ellipse((0.66, -4), 0.6, 0.8, color=cnred))
ax.add_patch(Ellipse((0.56, -4.2), 0.8, 0.6, color=cnred))
ax.add_patch(Ellipse((-0.66, -4), 0.6, 0.8, color=cnred))
ax.add_patch(Ellipse((-0.56, -4.2), 0.8, 0.6, color=cnred))

# Body
ax.add_patch(Ellipse((0, -3.25), 2.6, 1.9, color=cnred))
ax.add_patch(Ellipse((0, -3.25), 1.15, 1.15, color='white'))
ax.add_patch(Ellipse((0, -3.25), 1.25, 1.25, alpha=0.5, color=gold))
ax.add_patch(Ellipse((0, -3.25), 1, 1, color='white'))

# Scarf
tt = np.linspace(np.pi-0.35, np.pi+0.36, 100)
x = 3.25 * np.sin(tt)
y = 2.55 * np.cos(tt) + 0.1
ax.plot(x, y-0.02, color='black', linewidth=5)
ax.plot(x, y, color=gold, linewidth=6)
ax.add_patch(Rectangle((-1+0.02,-3.1-0.02),0.6,0.05,angle=70,color='black'))
ax.add_patch(Rectangle((-1.15+0.02,-3.045-0.02),0.25,0.05,angle=70,color='black'))
ax.add_patch(Rectangle((-1.3+0.02,-2.985-0.02),0.25,0.05,angle=70,color='black'))
ax.add_patch(Rectangle((-1.45+0.02,-2.925-0.02),0.25,0.05,angle=70,color='black'))
ax.add_patch(Rectangle((-0.91,-2.85),0.4,0.54,angle=70,color=gold))
ax.add_patch(Rectangle((-1,-3.1),0.25,0.05,angle=70,color=gold))
ax.add_patch(Rectangle((-1.15,-3.045),0.25,0.05,angle=70,color=gold))
ax.add_patch(Rectangle((-1.3,-2.985),0.25,0.05,angle=70,color=gold))
ax.add_patch(Rectangle((-1.45,-2.925),0.25,0.05,angle=70,color=gold))

# Pigion
def drawPigeon(x, y, sign=1): 
    ax.add_patch(Ellipse((x, y), 0.1, 0.2, angle=sign*45, color=gold))
    ax.add_patch(Ellipse((x+sign*0.1, y), 0.1, 0.2, angle=-sign*45, color=gold))
    ax.add_patch(Ellipse((x+sign*0.22, y+0.1), 0.2, 0.08, angle=sign*20, color=gold))
    ax.add_patch(Ellipse((x-sign*0.1, y+0.1), 0.1, 0.01, color=gold))
    ax.add_patch(Ellipse((x+sign*0.12, y-0.08), 0.05, 0.01, color=gold))
    ax.add_patch(Ellipse((x+sign*0.17, y-0.03), 0.05, 0.01, color=gold))

drawPigeon(0.25, 1.76)
drawPigeon(-0.25, 1.76, -1)
drawPigeon(1.5, 1.76)
drawPigeon(1.05, 1.76, -1)
drawPigeon(-1.05, 1.76)
drawPigeon(-1.5, 1.76, -1)

ax.add_patch(Rectangle((-2.2, 1.65), 4.4, 0.4, color=cnred))
ax.add_patch(Ellipse((-0.2, 2.4), 1.6, 0.8, color='w'))
ax.add_patch(Ellipse((-1.2, 2.35), 0.6, 1.6, angle=110, color='w'))
ax.add_patch(Ellipse((1.2, 2.3), 0.7, 1.4, angle=80, color='w'))


# 头顶的把手
tt1 = np.linspace(-np.pi/3.2, np.pi/3.2, 100) # 上方
tt2 = np.linspace(0, np.pi, 100) # 左右
ax.plot(0.4*np.sin(tt1), 0.4*np.cos(tt1)+2.9, color=gold, linewidth=1)
ax.plot(0.1*np.sin(tt2)+0.36, 0.1*np.cos(tt2)+3, color=gold, linewidth=1)
ax.plot(-0.1*np.sin(tt2)-0.36, 0.1*np.cos(tt2)+3, color=gold, linewidth=1)
    
# 脚上不知道啥装饰(长得像树？)
def drawTree(x, y): 
    ax.add_patch(Ellipse((x, y-0.05), 0.05, 0.2, color=gold))
    ax.add_patch(Ellipse((x, y+0.1), 0.05, 0.1, color=gold))
    ax.add_patch(Ellipse((x+0.1, y+0.07), 0.1, 0.1, color=gold))
    ax.add_patch(Ellipse((x-0.1, y+0.07), 0.1, 0.1, color=gold))
    ax.add_patch(Ellipse((x+0.2, y+0.02), 0.06, 0.06, color=gold))
    ax.add_patch(Ellipse((x-0.2, y+0.02), 0.06, 0.06, color=gold))
    ax.add_patch(Ellipse((x, y-0.15), 0.1, 0.01, color=gold))

drawTree(0.25,-4.2)
drawTree(0.75,-4.2)
drawTree(-0.25,-4.2)
drawTree(-0.75,-4.2)

    
# LOGO
ax.text(0, -3.4, 'BEIJING 2022', ha='center', fontsize=2)


# 设置图形的宽高比为1:1
ax.set_aspect('equal')

# 调整坐标轴范围以包含整个图形
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 4)

# 移除坐标轴
plt.axis('off')

# 显示图形
plt.show()    
```
